```python
# dropping of null value columns which are unwanted
# feature selection- here latitiude,longitude and location are not usefull for analysing
a = df.drop(['Historical Wards 2003-2015','Zip Codes','Community Areas','Census Tracts','Wards','Latitude','Longitude','Location'], axis=1)
a.head()
```

```python
# filling of null values with a specified name
a['Facility Type'] = a['Facility Type'].fillna('Food court')
```

```python
a = a.dropna(subset=['City','State','Zip'])
```

```python
a['Inspection Type'] = a['Inspection Type'].fillna('License')
```

```python
a['Risk'].value_counts()
```

```python
a['Risk'] = a['Risk'].fillna('no risk')
```

```python
a['AKA Name'] = a['AKA Name'].fillna('no name')
```

```python
a['AKA Name'].isnull().sum()
```

```python
a['License #'] = a['License #'].fillna(a['License #'].mode()[0])
```

```python
a.isnull().sum()
```

```python
a.head()
```

```python
# Function to fill nulls using mode for each risk category
def fill_nulls_by_risk(df):
    # Group by 'risk' and fill nulls in 'violations' using mode
    for risk_level in df['Risk'].unique():
        mode_value = df[df['Risk'] == risk_level]['Violations'].mode()
        df.loc[(df['Risk'] == risk_level) & (df['Violations'].isnull()), 'Violations'] = mode_value
    return df


df_1 = fill_nulls_by_risk(a)
print(df_1['Violations'].isnull().sum())
```

```python
final_df = df_1.dropna()
```
